LDA--linear discirminant analysis 线性判别分类\
思想——投影后类内方差最小，类间方差最大\
假设被降维的样本数据服从高斯分布；降维时采用了矩阵的特征分解思想

```python
# 导入数据集
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
print(X.shape,y.shape)
```

```python
# 使用PCA和LDA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X,y).transform(X)
```

```python
# 可视化
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False
print("first two components:%s"%str(pca.explained_variance_ratio_))
fig,axes = plt.subplots(1,2,figsize=(12,5))

colors = ['navy','turquoise','darkorange']
lw = 2
for color,i,target_name in zip(colors,[0,1,2],target_names):
    axes[0].scatter(X_r[y==i,0],X_r[y==i,1],color=color,alpha=0.8,lw=lw,label=target_name)
axes[0].legend(loc='best',shadow=False,scatterpoints=1)
axes[0].set_title("PCA of iris dataset")

for color,i,target_name in zip(colors,[1,2,3],target_names):
    axes[1].scatter(X_r2[y==i,0],X_r2[y==i,1],color=color,alpha=0.8,label=target_name)
axes[1].legend(loc='best',shadow=False,scatterpoints=1)
axes[1].set_title("LDA of iris dataset")
plt.show()
```
